```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
import os
from urllib.request import urlopen

class web_toon : 
    def __init__(self):
        # options
        self.options = webdriver.ChromeOptions()
        self.options.add_argument("User-Agent:Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.114 Safari/537.36")
        
        # call the chromedriver from local 
        self.chromedriver = 'C:\dev_project\python_crawling\chromedriver.exe'
        self.driver = webdriver.Chrome(self.chromedriver)

    # should be given the webtoon url and the episode number
    # be printed the image links
    def imgs_link(self, url, episode):
        
        try :
            # url with the episode number
            url = url + str(episode)
            
            # opens a window with url
            self.driver.get(url)
            
            # CSS_SELECTOR
            target = "body #wrap #container #content .section_cont.wide .view_area .wt_viewer"
            #comment_target =  "body ul.u_cbox_list li.u_cbox_comment div.u_cbox_text_wrap span.u_cbox_contents"
            
            # chose all news and the titles, if not exist, will be catched and thrown into exception
            all_imgs = WebDriverWait(toon.driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR , target)))
            
            # as List saved
            imgs = all_imgs.find_elements(By.CSS_SELECTOR,'img')
            

            # find out the image links
            self.imgs_url = []
            for i in range(len(imgs)):
                self.imgs_url.append(imgs[i].get_attribute("src"))
                print(imgs[i].get_attribute("src"))
            
        except:
            print("time out! not found!")

        finally:
            return
        
    

toon = web_toon()
toon.imgs_link("https://comic.naver.com/webtoon/detail.nhn?titleId=730694&no=",5)
```
